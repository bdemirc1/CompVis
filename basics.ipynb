### Displaying images

```python
#Show the image with matplotlib
# Matplotlib -> RGB, OPENCV -> BGR
plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
plt.show()
```

```python
def imshow(title= "", image=None):
    plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
    plt.title(title)
    plt.show()
```

```python
imshow("First image", image)
```

### Saving images

```python
# Or save it as png (portable network graphics)
cv2.imwrite("output.jpg", image)
```

```python
print('Height of Image: {} pixels'.format(int(image.shape[0])))
print('Width of Image: {} pixels'.format(int(image.shape[1])))
print('Depth of Image: {} colors components'.format(int(image.shape[2])))
```

### Grayscale

```python
imshow("Grayscale has no depth", image_gray)
```

```python
plt.imshow(image_gray, cmap="gray")
```

### Resizing

```python
rsized_img = cv2.resize(image, (1000, 400))
imshow("Resized image", rsized_img)
```

```python
# resizing with ratio
w_ratio = 0.5
h_ratio = 0.5
ratio_img = cv2.resize(image, (0,0), image, w_ratio, h_ratio)
imshow("Resized with ratio", ratio_img)
```

### Flipping images

```python
flip_img = cv2.flip(image, 1) # 0 for horizontal axis, upside down, -1 for both vertical and horizontal
imshow("Flipped through vertical axis", flip_img)
```
